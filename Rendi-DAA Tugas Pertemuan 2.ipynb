```python
# Fungsi untuk menghitung hasil kali semua elemen dalam list
def getKali(myList):
    hasil_kali = 1  # Inisialisasi dengan 1 karena 1 adalah elemen netral dalam perkalian
    for angka in myList:
        hasil_kali *= angka  # Perkalian setiap elemen
    return hasil_kali

# Misalkan kita punya list
myList = [2, 3, 4]

# Hasil perkalian semua elemen dalam myList
hasil_kali = getKali(myList)
print("Hasil perkalian seluruh angka:", hasil_kali)
```

```python
# Fungsi untuk membagi seluruh elemen dalam list
def getBagi(myList):
    if len(myList) == 0:
        return "List kosong tidak bisa dibagi."

    hasil_bagi = myList[0]  # Inisialisasi dengan elemen pertama
    for angka in myList[1:]:  # Loop mulai dari elemen kedua
        hasil_bagi /= angka  # Pembagian setiap elemen secara berurutan
    return hasil_bagi

# Misalkan kita punya list
myList = [250, 5, 10]

# Hasil pembagian seluruh elemen dalam myList
hasil_bagi = getBagi(myList)
print("Hasil bagi seluruh angka:", hasil_bagi)
```

```python
# Buatlah suatu fungsi untuk membagi 2 himpunan angka, Jika getSum untuk menjumlahkan angka!
# Fungsi untuk membagi dua himpunan angka secara berpasangan
def bagiDuaHimpunan(himpunan1, himpunan2):
    if len(himpunan1) != len(himpunan2):
        print("Himpunan harus memiliki panjang yang sama untuk pembagian yang tepat.")
        return None

    hasil_bagi = []
    for i in range(len(himpunan1)):
        if himpunan2[i] == 0:
            print("Pembagian dengan nol tidak diperbolehkan.")
            return None
        hasil_bagi.append(himpunan1[i] / himpunan2[i])  # Pembagian elemen secara berpasangan
    return hasil_bagi

# Misalkan kita punya dua himpunan angka
himpunan1 = [80, 40, 30]
himpunan2 = [5, 10, 15]

# Memanggil fungsi untuk membagi dua himpunan angka
hasil_bagi = bagiDuaHimpunan(himpunan1, himpunan2)
if hasil_bagi:
    print("Hasil bagi dari dua himpunan angka:", hasil_bagi)
```

```python

# Buatlah suatu fungsi untuk mengurangi 2 himpunan angka, jika getSum untuk menjumlahkan angka!
# Fungsi untuk mengurangi dua himpunan angka secara berpasangan
def kurangiDuaHimpunan(himpunan1, himpunan2):
    if len(himpunan1) != len(himpunan2):
        print("Himpunan harus memiliki panjang yang sama untuk pengurangan yang tepat.")
        return None

    hasil_kurang = []
    for i in range(len(himpunan1)):
        hasil_kurang.append(himpunan1[i] - himpunan2[i])  # Pengurangan elemen secara berpasangan
    return hasil_kurang

# Misalkan kita punya dua himpunan angka
himpunan1 = [50, 40, 30]
himpunan2 = [2, 4, 5]

# Memanggil fungsi untuk mengurangi dua himpunan angka
hasil_kurang = kurangiDuaHimpunan(himpunan1, himpunan2)
if hasil_kurang:
    print("Hasil pengurangan dari dua himpunan angka:", hasil_kurang)
```
